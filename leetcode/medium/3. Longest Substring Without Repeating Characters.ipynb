```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        lenght_dict = {}
        k = len(lenght_dict)
        mx = 0
        if len(s) == 0:
            return 0
        if s == ' ':
            return 1
        for i in s:
        
            if i not in lenght_dict:
                lenght_dict[i] = 1
                k = len(lenght_dict)
            else:
                lenght_dict = {}
                lenght_dict[i] = 1
            mx = max(k,mx)
            print(lenght_dict, k)
        return mx

sol = Solution()
sol.lengthOfLongestSubstring('dvdf')
```
