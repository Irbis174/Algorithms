```python
from typing import List

class Solution:
    def subarraySum(self, nums: List[int], k: int) -> int:
        left = 0
        sm = 0
        cnt = 0
        for right in range(len(nums)):
            while left <= right:
                print(sm, cnt, nums[left], nums[right], left, right)
                if nums[right] == 0:
                    right += 1
                sm += nums[left]
                if sm + nums[right] == k:
                    cnt += 1
                    sm -= nums[left]
                left += 1
        return cnt
    
sol = Solution()
sol.subarraySum([1,2,3], 3)
```
