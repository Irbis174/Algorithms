```python
from typing import List

class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        hash_dict = {}
        result = []
        for i in strs:
            anam = ''.join(sorted(i)) 
            if anam in hash_dict:
                hash_dict[anam].append(i)
            else:
                hash_dict[anam] = []
                hash_dict[anam].append(i)
            

        return list(hash_dict.values())
    
sol = Solution()
sol.groupAnagrams(["eat","tea","tan","ate","nat","bat"])
        
```
