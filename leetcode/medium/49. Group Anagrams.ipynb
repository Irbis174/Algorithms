```python
from typing import List

class Solution:
    def groupAnagrams(self, strs):
        hash_dict = {}
        arr = []

        for i in range(len(strs)):
            print(sorted(strs[i]))
            if sorted(strs[i]) not in hash_dict:
                hash_dict[sorted(strs[i])] = 1
            else:
                hash_dict[sorted(strs[i])] += 1

        return hash_dict    
    
sol = Solution()
sol.groupAnagrams(["eat","tea","tan","ate","nat","bat"])
```
