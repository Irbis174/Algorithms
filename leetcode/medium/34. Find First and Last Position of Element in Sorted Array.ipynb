```python
from typing import List
class Solution:
    def searchRange(self, nums: List[int], target: int) -> List[int]:
        if len(nums) == 0:
            return [-1,-1]
        if len(nums) == 1:
            if target == nums[0]:
                return [0,0]
            else:
                return [-1,-1]
        if len(nums) == 2:
            print(len(nums)//2)
        first = len(nums)
        last = len(nums)
        for i in range(len(nums) // 2):
            print(nums[i], nums[-i-1])
            if nums[i] == target:
                first = min(i,first)
            elif nums[-i-1] == target:
                last = min(abs(len(nums)-i), last)
            elif nums[i] > target or i == (len(nums) // 2):
                if nums[first] == target:
                    return [first,first]
                return [-1,-1]
        return [first, last]   

sol = Solution()
# sol.searchRange([5,7,7,8,8,10], 6)
sol.searchRange([1,3], 1)
```
