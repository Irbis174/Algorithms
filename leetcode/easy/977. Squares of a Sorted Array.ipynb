```python
from typing import List

class Solution:
    def sortedSquares(self, nums: List[int]) -> List[int]:
        left = 0
        right = len(nums) - 1
        p = len(nums) - 1
        arr = [0] * len(nums) 
        while left <= right:
            left_square = nums[left]**2
            right_square = nums[right]**2
            if left_square > right_square:
                arr[p] = left_square
                left += 1
            else:
                arr[p] = right_square
                right -= 1
            p -= 1

        return arr
            
sol = Solution()
sol.sortedSquares([-7,-3,2,3,11])
```
