```python
class Solution:
    def firstUniqChar(self, s: str) -> int:
        hash_dict = {}
        for i in range(len(s)):
            if s[i] not in hash_dict:
                hash_dict[s[i]] = i
            else:
                hash_dict[s[i]] = 'a'
        print(hash_dict)
        
        for i in hash_dict:
            print(hash_dict[i])
            if hash_dict[i] != 'a':
                return hash_dict[i]
        return -1
    
sol = Solution()
sol.firstUniqChar('leetcode')
```
