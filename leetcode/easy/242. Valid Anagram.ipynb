```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) != len(t):
            return False
        
        hash_dict = {}
        
        for i in range(len(s)):
            if s[i] not in hash_dict:
                hash_dict[s[i]] = 1
            else:
                hash_dict[s[i]] += 1

        for i in t:
            if i not in hash_dict:
                return False
            hash_dict[i] -= 1
            if hash_dict[i] == 0:
                del hash_dict[i]

        return len(hash_dict) == 0
```
