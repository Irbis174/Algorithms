```python
class Solution:
    def canConstruct(self, ransomNote: str, magazine: str) -> bool:
        hash_dict_1 = {}
        hash_dict_2 = {}
        
        for i in magazine:
            hash_dict_1[i] = 1 + hash_dict_1.get(i,0)
            
        for i in ransomNote:
            hash_dict_2[i] = 1 + hash_dict_2.get(i,0)
        
        for i in hash_dict_2:
            if i not in hash_dict_1 or hash_dict_2[i] > hash_dict_1[i]:
                return False
        return True
            
sol = Solution()
sol.canConstruct('a','b')
sol.canConstruct('aa','ab')
sol.canConstruct('aa','aab')
sol.canConstruct('aa')
```

```python
class Solution:
    def canConstruct(self, ransomNote: str, magazine: str) -> bool:
        hash_dict = {}
        #length = len(ransomNote)
        for i in ransomNote:
            hash_dict[i] = 1 + hash_dict.get(i, 0)
        print(hash_dict)
        for i in magazine:
            if i in hash_dict:
                hash_dict[i] -= 1
                if hash_dict[i] == 0:
                    del hash_dict[i]
                magazine = magazine.replace(i,'',1)
                if len(magazine) == 0:
                    return True
        return False
               

sol = Solution()
sol.canConstruct('bg', 'efjbdfbdgfjhhaiigfhbaejahgfbbgbjagbddfgdiaigdadhcfcj')
sol.canConstruct('aa','ab')
sol.canConstruct('aa','aab')
```
