```python
class Solution:
    def canConstruct(self, ransomNote: str, magazine: str) -> bool:
        hash_dict_1 = {}
        hash_dict_2 = {}
        for i in range(len(magazine)):
            hash_dict_1[magazine[i]] = 1 + hash_dict_1.get(magazine[i], 0)
            if i <= len(ransomNote):
                hash_dict_2[ransomNote[i]] = 1 + hash_dict_2.get(ransomNote[i], 0)

        for i in hash_dict_2:
            print(i in hash_dict_2.keys())
            if i in hash_dict_1.keys():
                if (hash_dict_2[i] < hash_dict_1[i]):
                    return False
        
        return True
    

sol = Solution()
sol.canConstruct('a', 'b')
sol.canConstruct('aa', 'ab')
sol.canConstruct('aa', 'aab')
        
```
