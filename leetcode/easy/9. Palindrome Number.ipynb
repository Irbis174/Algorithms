```python
from dataclasses import dataclass

@dataclass
class Solution:
    def isPalindrome(self, x: int) -> bool:
        if x < 0:
            return False
        if x == 0:
            return True
        st = str(x)
        for i in range(len(st) // 2 + 1):
            if st[i] != st[-i-1]:
                return False

        return True
    
sol = Solution()
sol.isPalindrome(x = 121)
sol.isPalindrome(x = 123)
sol.isPalindrome(x = 1221)
```
