```python
def cnt_pairs_with_diff_gtk(nums, k):
    cnt_pairs = 0
    last = 0
    for first in range(len(nums)):
        while last < len(nums) and nums[last] - nums[first] <= k:
            last += 1
        cnt_pairs += len(nums) - last
    return cnt_pairs
```

```python
def best_team_sum(players):
    best_sum = 0
    now_sum = 0
    last = 0
    for first in range(len(players)):
        while last < len(players) and (last == first or players[first] + players[first + 1] >= players[last]):
            now_sum += players[last]
            last += 1
        best_sum = max(best_sum, now_sum)
        now_sum -= players[first]
    return best_sum
```

```python
def merge(arr1, arr2):
    n = len(arr1)
    m = len(arr2)

    merged = [0] * (n+m)
    first1 = first2 = 0

    for i in range(n+m):
        if first1 != n and first2 == m or arr1[first1] <= arr2[first2]:
            merged[i] = arr1[first1]
            first1 +=1
        else:
            merged[i] = arr1[first2]
            first2 += 1 
    
    return merged
```

```python
n, k = map(int, input().split())
s = input()
d = dict()
l = r = bestl = bestr = 0

while r < n:
    if d.get(s[r], 0) < k:
        if r - l > bestr - bestl:
            bestr, bestl = r, l
        d[s[r]] = d.get(s[r], 0) + 1
        r += 1
    else:
        while l < n and d.get(s[r], 0) >= k:
            d[s[l]] -= 1
            l += 1
print(bestr - bestl + 1, bestl + 1)
```
