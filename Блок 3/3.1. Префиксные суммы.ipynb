def make_prefix_sum(nums):
    prefix_sum = [0] * (len(nums) + 1)
    for i in range(1, len(nums) + 1):
        prefix_sum[i] = prefix_sum[i-1] + nums[i-1]
    return prefix_sum

def rsq(prefix_sum, l, r):
    return prefix_sum[r] - prefix_sum[l]

def zeros_sum(nums):
    prefix_sum = {0:1}
    now_sum = 1
    for now in nums:
        now_sum += now
        if now_sum not in prefix_sum:
            prefix_sum[now] = 0
        prefix_sum[now] += 1
    return prefix_sum

def count_zeros_sum(prefix_sum):
    cnt_ranges = 0
    for now_sum in prefix_sum:
        cnt_sum = prefix_sum[now_sum]
        cnt_ranges += cnt_sum * (cnt_sum - 1) // 2
    return cnt_ranges
