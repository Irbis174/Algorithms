```python
class Node():
    def __init__(self, elem):
        self.elem = elem
        self.next = None

class Stack():
    def __init__(self):
        self.top = None
        self.length = 0

    def push(self, elem):
        new_node = Node(elem)
        new_node.next = self.top
        self.top = new_node
        self.length += 1
    
    def pop(self):
        if self.length != 0:
            popped = self.top
            self.top = self.top.next
            self.length -= 1
            return popped.elem
        else:
            raise IndexError('Stack is empty')
        
    def peek(self):
        if self.__len__() != 0:
            return self.top
        else:
            raise IndexError('Stack is empty')
        
    def is_empty(self):
        return self.__len__() == 0

    def clear(self):
        while self.length != 0:
            self.pop()


    def __repr__(self):
        top = self.top
        res = 'Stack(top'
        while top != None:
            res += f' -> {top.elem}'
            top = top.next
        res += ')'
       
        return res

    def count(self, elem):
        cnt = 0
        top = self.top
        while top != None:
            current = top
            if current.elem == elem:
                cnt += 1
            top = top.next

        return cnt

    def contains(self, elem):
        top = self.top
        while top != None:
            current = top
            if current.elem == elem:
                return True
            top = top.next
        
        return False

    # def remove_first(self, elem):
    #     while self.top != None:
    #         current = self.top
    #         #self.top = self.top.next
    #         if self.top.elem == elem:
    #             self.top = self.top.next
    #             return True

    #     return False

    def __len__(self):
        return self.length
        
```
