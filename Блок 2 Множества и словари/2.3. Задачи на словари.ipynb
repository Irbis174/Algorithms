```python
n = int(input())
max_heigth = {}
for i in range(n):
    w, h = map(int, input().split())
    max_heigth[w] = max(max_heigth.get(w,0),h)
ans = 0
for wight in max_heigth:
    ans += max_heigth[wight]
print(ans)
```

```python
import sys


def main():
    n = int(input())
    words = {}
    for i in range(n):
        words_tpl = tuple(map(str, input().split()))
        a, b = words_tpl[0], words_tpl[1]
        words[a] = b
        words[b] = a

    word = input()
    if word in words:
        return words[word]


if __name__ == '__main__':
    print(main())
```

```python
import sys


def main():
    clients = {}
    with open('input.txt', 'r', encoding='utf8') as f:
        for line in f:
            who, whats, cnt = line.strip().split()
            if who not in clients:
                clients[who] = {}
            if whats not in clients[who]:
                clients[who][whats] = 0
            clients[who][whats] += int(cnt)
    ans = []
    for who in sorted(clients):
        ans.append(f'{who}:')
        for whats in sorted(clients[who]):
            ans.append(whats + ' ' + str(clients[who][whats]))
    
    print('\n'.join(ans))




if __name__ == '__main__':
    main()
```

```python
import sys


def main():
    bank = {}
    ans = []
    with open('input.txt', 'r', encoding='utf8') as f:
        for line in f:
            st = line.strip().split()
            if st[0] == 'DEPOSIT':
                if st[1] not in bank:
                    bank[st[1]] = int(st[2])
                else:
                    bank[st[1]] += int(st[2])
            elif st[0] == 'TRANSFER':
                if st[1] not in bank:
                    bank[st[1]] = -int(st[3])
                else:
                    bank[st[1]] -= int(st[3])
                if st[2] not in bank: 
                    bank[st[2]] = int(st[3])
                else:
                    bank[st[2]] += int(st[3])
            elif st[0] == 'INCOME':
                for i in bank.keys():
                    if bank[i] >= 0:
                        bank[i] += int((int(st[1]) / 100 * bank[i]))
            elif st[0] == 'WITHDRAW':
                if st[1] not in bank:
                    bank[st[1]] = -int(st[2])
                else:
                    bank[st[1]] -= int(st[2])
            elif st[0] == 'BALANCE':
                if st[1] not in bank:
                    ans.append('ERROR')
                else:
                    ans.append(bank[st[1]])
    print('\n'.join(str(x) for x in ans)) 



if __name__ == '__main__':
    main()
```
