```python
import sys


def main():
    palindrom = {}
    n = int(input())
    word = input()
    for i in word:
        if i not in palindrom:
            palindrom[i] = 1
        else:
            palindrom[i] += 1
    sorted_palindrom = sorted(palindrom)
    for i in sorted_palindrom:
        print(i * (palindrom[i] // 2), end='')

    for i in sorted_palindrom:
        if palindrom[i] % 2 == 1:
            print(i, end='')
            break
    sorted_palindrom.reverse()
    for i in sorted_palindrom:
        print(i * (palindrom[i] // 2), end='')

if __name__ == '__main__':
    main()
```
