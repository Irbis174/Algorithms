```python
import sys


def main():
    nums = list(map(int, input().split()))
    x = int(input())
    
    prev_nums = set()
    for now_num in nums:
        if x - now_num in prev_nums:
            return now_num, x - now_num
        prev_nums.add(now_num)
    return 0, 0 

if __name__ == '__main__':
    print(main())
```

```python
import sys


def main():
    n = int(input())
    used = set()
    for i in range(n):
        tpl = tuple(map(int, input().split()))
        a, b = tpl[0], tpl[1]
        if a >= 0 and b >= 0 and a + b == n - 1:
            used.add(tpl)
    return len(used)


if __name__ == '__main__':
    print(main())
```
