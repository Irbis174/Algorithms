```python
n = int(input())
game = {}
for i in range(n):
    name = input()
    game[name] = 0
m = int(input())
score_last = (0,0)
for i in range(m):
    tpl = tuple(map(str, input().split()))
    score, name = tpl
    score = tuple(map(int, score.split(':')))
    if score[0] != score_last[0]:
        game[name] += (score[0] - score_last[0])
    elif score[1] != score_last[1]:
        game[name] += (score[1] - score_last[1])
    score_last = score

mx = 0
name = ''
for i in game:
    if game[i] > mx:
        name = i
        mx = max(game[i], mx)

print(name, mx, end=' ')
```
