# Покраска деревьев


```python
import sys


def main():
    arr_1 = list(map(int, input().split()))
    arr_2 = list(map(int, input().split()))
    
    n_1 = ((arr_1[0]+arr_1[1]) - (arr_1[0]-arr_1[1])) + 1
    n_2 = ((arr_2[0]+arr_2[1]) - (arr_2[0]-arr_2[1])) + 1

    while arr_1[1] != 0 and arr_2[1] != 0:
        if arr_1[1] > arr_2[1]:
            arr_1[1] -= arr_2[1]
        else:
            arr_2[1] -= arr_1[1]
    
    return n_1 + n_2 - arr_1[1]

if __name__ == '__main__':
    print(main())
```
